### Question 1: What is Data Loading in Power BI and why is it considered the first step of analysis?

**Data Loading** in Power BI is the process of **importing or connecting data from various data sources** (such as Excel, CSV files, SQL Server, Oracle, cloud services, or web APIs) into Power BI Desktop for analysis and reporting.

It is considered the **first step of analysis** because Power BI must access the data before it can be cleaned, transformed, modeled, visualized, or analyzed.

### Why is Data Loading the First Step of Analysis?

1. **Imports Data into Power BI**

   * It brings data from one or more sources into Power BI for further processing.

2. **Enables Data Cleaning**

   * Once loaded, data can be cleaned using **Power Query** by removing duplicates, handling missing values, and correcting data types.

3. **Supports Data Modeling**

   * Loaded data can be used to create relationships between tables and build a data model.

4. **Creates Reports and Dashboards**

   * After loading and preparing the data, users can create charts, graphs, maps, KPIs, and interactive dashboards.

5. **Provides Business Insights**

   * Accurate and complete data is the foundation for reliable analysis and informed decision-making.

### Data Loading Workflow

```text
Data Source
     ↓
Load Data into Power BI
     ↓
Clean & Transform Data (Power Query)
     ↓
Create Data Model
     ↓
Build Reports & Dashboards
     ↓
Analyze Business Insights
```

### Examples of Data Sources

* Excel Workbook
* CSV/Text Files
* SQL Server
* Oracle Database
* MySQL
* SharePoint
* Azure SQL Database
* Web APIs

### Conclusion

**Data Loading** is the process of importing or connecting data into Power BI from different sources. It is the **first step of analysis** because all subsequent tasks—data cleaning, transformation, modeling, visualization, and reporting—depend on having the data available in Power BI.


### Question 2: Explain the difference between “Load” and “Transform Data” in Power BI.

In Power BI, after selecting a data source, you are typically given two options: **Load** and **Transform Data**. They serve different purposes.

| **Load**                                                        | **Transform Data**                                                                                                        |
| --------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------- |
| Imports the data directly into Power BI without making changes. | Opens the **Power Query Editor** to clean and modify the data before loading.                                             |
| Quick and suitable when the data is already clean.              | Used when the data requires preparation before analysis.                                                                  |
| Saves time because no data transformation is performed.         | Allows you to remove duplicates, handle missing values, change data types, filter rows, merge tables, split columns, etc. |
| Data is immediately available for creating reports.             | Data is loaded only after clicking **Close & Apply**.                                                                     |

### Load

* Imports the selected dataset directly into Power BI.
* Best used when the data is already accurate and properly formatted.
* No cleaning or transformation is performed before loading.

**Example:** Loading a clean Excel file directly to create reports.

### Transform Data

* Opens the **Power Query Editor** before loading the data.
* Used to clean, reshape, and prepare the data for analysis.
* Common transformations include:

  * Removing duplicate rows
  * Handling NULL values
  * Changing data types
  * Renaming columns
  * Splitting or merging columns
  * Filtering unnecessary rows

**Example:** Cleaning a CSV file that contains missing values and inconsistent date formats before importing it into Power BI.

### Summary

* **Load** → Imports data directly into Power BI without any modifications.
* **Transform Data** → Opens Power Query to clean, transform, and prepare the data before loading.

**Recommendation:** Use **Load** when your data is already clean. Use **Transform Data** when the dataset needs cleaning or restructuring before analysis.


### Question 3: What is a Fact Table and a Dimension Table? Give examples from the dataset.

A **Fact Table** and a **Dimension Table** are the two main types of tables used in a **Power BI data model** (Star Schema).

| Table Type          | Definition                                                                           | Example                                                                |
| ------------------- | ------------------------------------------------------------------------------------ | ---------------------------------------------------------------------- |
| **Fact Table**      | Stores measurable business data (facts) such as sales, revenue, quantity, or profit. | COVID-19 statistics like confirmed cases, deaths, and recovered cases. |
| **Dimension Table** | Stores descriptive information that provides context to the facts.                   | Country, Date, State/Province, Region, etc.                            |

### 1. Fact Table

A **Fact Table** contains:

* Numeric values (measures)
* Foreign keys linked to dimension tables
* Transactional or measurable data

**Examples from the Coronavirus dataset:**

* Confirmed Cases
* Deaths
* Recovered Cases
* Active Cases (if available)

| Date       | Country | Confirmed | Deaths | Recovered |
| ---------- | ------- | --------: | -----: | --------: |
| 01-04-2020 | India   |      1998 |     58 |       148 |

Here, **Confirmed**, **Deaths**, and **Recovered** are **facts (measures)**.

---

### 2. Dimension Table

A **Dimension Table** contains descriptive attributes used to categorize or filter the facts.

**Examples from the Coronavirus dataset:**

* Country
* State/Province
* Region
* Date

| Country | Region        |
| ------- | ------------- |
| India   | Asia          |
| USA     | North America |
| Italy   | Europe        |

Or:

| Date       | Month | Year |
| ---------- | ----- | ---- |
| 01-04-2020 | April | 2020 |

These tables help users filter and analyze the fact data.

---

### Example Star Schema

```text
           Date Dimension
                 │
                 │
Country Dimension ─── Fact Table ─── Region Dimension
                 │
                 │
         State Dimension
```

The **Fact Table** contains:

* Confirmed Cases
* Deaths
* Recovered Cases

The **Dimension Tables** contain:

* Date
* Country
* State/Province
* Region

### Conclusion

* **Fact Table:** Contains measurable business data such as **Confirmed Cases, Deaths, and Recovered Cases**.
* **Dimension Table:** Contains descriptive information such as **Country, Date, State/Province, and Region**, which is used to filter and analyze the facts.

Together, Fact and Dimension Tables form the basis of an efficient **Power BI data model** for reporting and analysis.


### Question 4: Why is Star Schema preferred over Snowflake Schema in Power BI?

**Star Schema** is preferred over **Snowflake Schema** in Power BI because it is **simpler, faster, and easier to understand**. It provides better query performance and makes report creation more efficient.

### Difference Between Star Schema and Snowflake Schema

| Star Schema                                                | Snowflake Schema                                                      |
| ---------------------------------------------------------- | --------------------------------------------------------------------- |
| Dimension tables are directly connected to the Fact table. | Dimension tables are further normalized into multiple related tables. |
| Simple and easy to design.                                 | More complex to design and maintain.                                  |
| Faster query performance.                                  | Slower queries due to multiple table joins.                           |
| Fewer relationships between tables.                        | More relationships and joins between tables.                          |
| Preferred for Power BI reports and dashboards.             | Used mainly in highly normalized data warehouses.                     |

### Why Star Schema is Preferred in Power BI

1. **Better Performance**

   * Requires fewer table joins, resulting in faster report queries and dashboard performance.

2. **Simple Data Model**

   * Easy to understand, develop, and maintain.
   * Business users can work with the model more easily.

3. **Improved DAX Performance**

   * DAX measures and calculations execute more efficiently with a simpler schema.

4. **Easier Report Creation**

   * Relationships are straightforward, making it easier to create charts, slicers, and dashboards.

5. **Optimized for Analytics**

   * Power BI is designed to work efficiently with star schemas, making data analysis and aggregation faster.

### Example

**Star Schema**

```text
         Date
           │
           │
Customer ─ Sales Fact ─ Product
           │
           │
         Region
```

* One central **Fact Table (Sales Fact)**.
* Multiple **Dimension Tables** (Date, Customer, Product, Region) connected directly to it.

### Conclusion

Power BI prefers the **Star Schema** because it:

* Offers **better query performance**.
* Uses **fewer joins**, making reports faster.
* Is **easier to understand and maintain**.
* Improves **DAX calculations** and report development.

For these reasons, the **Star Schema is the recommended data modeling approach in Power BI**.


### Question 5: Identify and remove duplicate records based on Date, Country, and State.

To identify and remove duplicate records in **Power BI** using **Power Query**, follow these steps:

### Steps

1. Open **Power BI Desktop**.
2. Click **Home → Transform Data** to open **Power Query Editor**.
3. Press **Ctrl** and select the following columns:

   * **Date**
   * **Country**
   * **State**
4. Go to the **Home** tab.
5. Click **Remove Rows** → **Remove Duplicates**.
6. Power Query will keep the **first occurrence** of each unique combination of **Date, Country, and State** and remove any additional duplicate records.
7. Click **Home → Close & Apply** to save the changes and load the cleaned data.

### Example

**Before Removing Duplicates**

| Date       | Country | State       | Confirmed |
| ---------- | ------- | ----------- | --------: |
| 01-04-2020 | India   | Maharashtra |       320 |
| 01-04-2020 | India   | Maharashtra |       320 |
| 02-04-2020 | India   | Delhi       |       219 |
| 03-04-2020 | USA     | New York    |     83712 |

**After Removing Duplicates**

| Date       | Country | State       | Confirmed |
| ---------- | ------- | ----------- | --------: |
| 01-04-2020 | India   | Maharashtra |       320 |
| 02-04-2020 | India   | Delhi       |       219 |
| 03-04-2020 | USA     | New York    |     83712 |

### Result

* Duplicate records based on the combination of **Date + Country + State** are removed.
* Only one unique record for each combination is retained.
* The dataset becomes cleaner and more reliable for analysis and reporting.


### Question 6: Identify and replace null values in `Vaccination_State`

To identify and replace **NULL** values in the **`Vaccination_State`** column using **Power Query** in Power BI, follow these steps:

### Steps to Identify NULL Values

1. Open **Power BI Desktop**.
2. Click **Home → Transform Data** to open **Power Query Editor**.
3. Select the **`Vaccination_State`** column.
4. Click the **filter drop-down** in the column header.
5. Look for **(null)** in the list of values.

   * If **(null)** appears, those records contain missing values.

### Steps to Replace NULL Values

1. Select the **`Vaccination_State`** column.
2. Go to the **Transform** tab.
3. Click **Replace Values**.
4. In the dialog box:

   * **Value to Find:** `null`
   * **Replace With:** `Not Vaccinated` *(or another appropriate value such as `Unknown`, depending on business requirements)*
5. Click **OK**.
6. Click **Home → Close & Apply** to save the changes.

### Example

**Before Replacement**

| Country | State       | Vaccination_State |
| ------- | ----------- | ----------------- |
| India   | Odisha      | Vaccinated        |
| India   | Delhi       | **null**          |
| USA     | Texas       | Fully Vaccinated  |
| India   | Maharashtra | **null**          |

**After Replacement**

| Country | State       | Vaccination_State  |
| ------- | ----------- | ------------------ |
| India   | Odisha      | Vaccinated         |
| India   | Delhi       | **Not Vaccinated** |
| USA     | Texas       | Fully Vaccinated   |
| India   | Maharashtra | **Not Vaccinated** |

### Result

* All **NULL** values in the `Vaccination_State` column are replaced with a meaningful value.
* The dataset becomes more complete and suitable for analysis, filtering, and reporting in Power BI.


### Question 7: Create a new column to calculate Recovery Rate.

The **Recovery Rate** represents the percentage of confirmed cases that have recovered.

### Formula

**Recovery Rate (%) = (Recovered Cases ÷ Confirmed Cases) × 100**

### Steps in Power BI (Power Query)

1. Open **Power BI Desktop**.
2. Click **Home → Transform Data**.
3. In **Power Query Editor**, go to the **Add Column** tab.
4. Click **Custom Column**.
5. Enter the column name: **Recovery Rate**.
6. Enter the following formula:

```powerquery
= ([Recovered] / [Confirmed]) * 100
```

7. Click **OK**.
8. Change the data type of the new column to **Decimal Number** or **Percentage**.
9. Click **Home → Close & Apply**.

### Alternative: Create a DAX Calculated Column

If you want to create the column after loading the data:

```DAX
Recovery Rate = DIVIDE('COVID_Data'[Recovered], 'COVID_Data'[Confirmed], 0) * 100
```

> `DIVIDE()` is recommended because it safely handles cases where **Confirmed = 0**, preventing division-by-zero errors.

### Example

| Country | Confirmed | Recovered | Recovery Rate |
| ------- | --------: | --------: | ------------: |
| India   |    10,000 |     8,500 |           85% |
| USA     |    20,000 |    15,000 |           75% |
| Brazil  |     5,000 |     4,250 |           85% |

### Result

A new column **Recovery Rate** is created, showing the percentage of recovered cases for each record. This metric can be used in Power BI reports, KPIs, and dashboards to analyze recovery performance across countries or regions.


### Question 8: Create a summarized table showing total confirmed cases by Country.

To create a summarized table showing the **total confirmed cases by Country** in Power BI, you can use either **Power Query** or **DAX**.

### Method 1: Using DAX (Recommended)

1. Open **Power BI Desktop**.
2. Go to the **Modeling** tab.
3. Click **New Table**.
4. Enter the following DAX formula:

```DAX
Country_Summary =
SUMMARIZE(
    COVID_Data,
    COVID_Data[Country],
    "Total Confirmed", SUM(COVID_Data[Confirmed])
)
```

> Replace **`COVID_Data`** with the actual name of your table.

---

### Method 2: Using Power Query

1. Click **Home → Transform Data**.
2. In **Power Query Editor**, select the dataset.
3. Select the **Country** column.
4. Go to **Transform → Group By**.
5. Configure the settings:

   * **Group By:** Country
   * **New Column Name:** Total Confirmed
   * **Operation:** Sum
   * **Column:** Confirmed
6. Click **OK**.
7. Click **Close & Apply**.

---

### Example Output

| Country | Total Confirmed |
| ------- | --------------: |
| India   |         125,000 |
| USA     |         250,000 |
| Brazil  |         180,000 |
| Italy   |          95,000 |

---

### Result

The summarized table displays the **total confirmed COVID-19 cases for each country**, making it easy to:

* Compare countries.
* Create bar charts, maps, and dashboards.
* Analyze country-wise trends in confirmed cases.
